```python
## get 100000 tweets

tweets = []

old_len = 0

while len(tweets) < 100:
    if len(tweets) == 0:
        tweets = api.search(q=top_trending_hashtag, count=100)
    else:
        lowest_tweet_id = tweets[len(tweets)-100]._json['id']
        for tweet in tweets:
            tweet_id = tweet._json['id']
            if tweet_id < lowest_tweet_id:
                lowest_tweet_id = tweet_id
        lowest_tweet_id -= 1
        tweets += api.search(q=top_trending_hashtag, count=100, max_id=lowest_tweet_id)
    print(len(tweets))
    if len(tweets) == old_len:
        break
    old_len = len(tweets)
        
```

```python
#fetch languages

url = "https://api.twitter.com/1.1/help/languages.json"
languages = requests.get(url, auth=auth).json()

def get_lang_text(code):
    for lang in languages:
        if lang['code'] == code:
            return lang['name']
    return "Unknown"
# search_resp = requests.get(, head)
# search_resp
```

```python
# for language aspect

lang_cnt={}

for tweet in tweets:
    if tweet.lang not in lang_cnt:
        lang_cnt[tweet.lang] = 0
    lang_cnt[tweet.lang]+=1
```

```python
print(lang_cnt)
for code in lang_cnt.keys():
    print(get_lang_text(code))
```
